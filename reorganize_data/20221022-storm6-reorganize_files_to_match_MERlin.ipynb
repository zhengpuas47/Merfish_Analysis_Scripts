# Resave STORM6 raw data into MERlin format

* Data from: \\10.245.74.158\Chromatin_NAS_0\20220304-P_brain_M1_nonclear_adaptors

```python
import os,glob,sys,time
import shutil
import numpy as np
sys.path.append(r"E:\Users\puzheng_new\Documents")
import ImageAnalysis3 as ia3
```

```python
data_folder = r'\\10.245.74.158\Chromatin_NAS_8\20221022-hM1_hM1_Cx28_sample1'

fds, fovs = ia3.get_img_info.get_folders(data_folder)
```

```python
#merfish_data_folder = r'I:\MERFISH_Data'
merfish_data_folder = r'F:\MERFISH_Data'

# target folder:
target_folder = os.path.join(merfish_data_folder, r'20221022-hM1_hM1')
#target_folder = os.path.join(merfish_data_folder, r'20221003-hMTG_hM1_test')

if not os.path.isdir(target_folder):
    print(f"creating folder: {target_folder}")
    os.makedirs(target_folder)
ref_image_type = 'Epi-750-647-488-405-s13'
image_type = 'Epi-750-647-488-s13'
polyt_type = 'Epi-750-488-405-s50'
```

```python
%%time
fov_ids = np.arange(len(fovs))

ref_round = 0
polyt_round = 11

#fov_ids = np.arange(2)
overwrite = False
remove_source = False

# loop through fovs
for _fov_id in fov_ids:
    
    _save_id = _fov_id 

    # files
    for _fd in fds[:]:
        _files = [os.path.join(_fd, _fl) for _fl in os.listdir(_fd) if fovs[_fov_id].split(os.extsep)[0] in _fl]
        _round = int(os.path.basename(_fd).split('H')[1].split('M')[0])
        if _round == ref_round:
            _image_type = ref_image_type
        elif _round == polyt_round:
            _image_type = polyt_type
        else:
            _image_type = image_type
        _image_round = _round
        # target file
        _target_files = [os.path.join(target_folder, f"{_image_type}_{_save_id}_{_image_round}{os.extsep}{_fl.split(os.extsep)[-1]}") 
                         for _fl in _files]
        for _fl, _tar_fl in zip(_files, _target_files):
            if remove_source:
                print(f"move {_fl} to {_tar_fl}")
                shutil.move(_fl, _tar_fl)
            else:
                if not os.path.isfile(_tar_fl) or overwrite:
                    print(f"copy {_fl} to {_tar_fl}")
                    shutil.copyfile(_fl, _tar_fl)
                else:
                    print(f"{_tar_fl} already exist, skip")
    if 'test' in target_folder and _fov_id >= 20:              
        break
```
